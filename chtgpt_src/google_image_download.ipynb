```python
import os
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time

# 크롬 드라이버 설정
driver = webdriver.Chrome()

# 이미지를 다운로드할 디렉토리 생성
download_dir = "image_downloads"
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

# Google 이미지 검색 페이지 열기
search_query = "블랙핑크"
driver.get(f"https://www.google.com/search?q={search_query}&tbm=isch")

# 이미지 URL을 추출하여 다운로드
image_count = 0
while image_count < 100:
    # 페이지 소스를 가져오고 BeautifulSoup로 파싱
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    
    # 이미지 URL 추출
    images = soup.find_all('img', class_='rg_i Q4LuWd')
    for image in images:
        image_url = image.get('src')
        
        # 이미지 URL이 유효하면 다운로드
        if image_url and image_url.startswith("http"):
            image_count += 1
            # image_extension = image_url.split('.')[-1]
            
            # 파일 이름에서 사용할 수 없는 문자 제거
            image_filename = f"image_{image_count}.jpg"
            # image_filename = "".join(x for x in image_filename if x.isalnum() or x in ['.', '_'])
        
            image_filename = os.path.join(download_dir, image_filename)
            
            response = requests.get(image_url)
            with open(image_filename, 'wb') as img_file:
                img_file.write(response.content)
            print(f"다운로드: {image_filename}")
    
    # 스크롤 다운
    driver.execute_script("window.scrollBy(0, 1000);")
    time.sleep(1)

# 브라우저 종료
driver.close()
```
